# Tuning / Optimization Cluster

```python
from sklearn.cluster import KMeans

wcss = []

clusters = range(1,15)

for k in range(1,15):
    kmeans2 = KMeans(n_clusters=k)
    kmeans2.fit(df[['score','income','Gender','Age']])
    wcss.append(kmeans2.inertia_)  
```

```python
wcss
```

```python
plt.plot(clusters , wcss)
```
