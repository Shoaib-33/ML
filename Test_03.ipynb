```python
for i in df1.columns:
    if is_numeric_dtype(df1[i]):
        continue
    else:
        one = pd.get_dummies(df1[i],drop_first=True,prefix='New')
        df1 = pd.concat([df1,one],axis=1).drop(i,axis=1)


  
    
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
for col in df3.columns:
   if is_numeric_dtype(df3[col]):
        continue
   else :     
     unique = df3[col].unique()
     df3[col] = OrdinalEncoder(categories=[unique]).fit_transform(df3[[col]])
```
